```python
#code to create my own reduce function that acts same as built-in reduce function
def myreduce(func,seq):
    output=seq[0]
    for i in seq[1:]:
        output=func(output,i)
    return output
```

```python
#testing of myreduce function
l=[2,4,6,8]
#add all elements in the list
myreduce(lambda a,b:a+b,l)
```

```python
#code to create my own filter function that acts same as built-in filter function
a=[]
def myfilter(func,seq):
    for i in seq:
        if func(i):
            a.append(i)
    return a
```

```python
#creating my own function to filter out all strings from the list l
l=[2,3,5,'prab','karan','navin']
b=[]
def test(a):
    if type(a)==str:
        b.append(a)
    return b
        
```

```python
#testing of myfilter function 
myfilter(test,l)
```
