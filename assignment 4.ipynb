```python
#To find area of triangle 
class tri:
    def __init__(self,a,b,c):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)
a= int(input("a="))
b= int(input("b="))
c= int(input("c="))

class triangle(tri):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)

    def area(self):
        s = (a + b + c) / 2
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5        

t = triangle(a,b,c)
print("area : {}".format(t.area()))
```

```python
#function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.
def filter_long_words(list_words,integer_n):
    list=[]
    for i in list_words:
        if len(i)>integer_n:
            list.append(i)
    return list
            
```

```python
filter_long_words(['banana', 'apple', 'orange', 'elephant', 'raspberry'],4)
```

```python
#program to find weather a character is vowel or not and return true if it is vowel else return false
def find_vowel(a):
    vowels=['a','e','i','o','u']
    if a in vowels:
            
        return True
    
    else:
        
        return False
```

```python
find_vowel('i')
```

```python
#program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.
def word_to_integer(a):
    l=[]
    for i in a:
        l.append(len(i))
    return l
        
```

```python
word_to_integer(['karan','prabhakar','mahabharath'])
```
